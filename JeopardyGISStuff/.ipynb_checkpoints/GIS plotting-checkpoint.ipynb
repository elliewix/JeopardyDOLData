```python
import csv
from shapely.geometry import Point, mapping
from fiona import collection 
```

```python
with open('latlon.csv', 'rt') as fin:
    data = csv.reader(fin)
    headers = next(data)
    latlon = [{'city': row[0].decode('utf-8'), 'lat': row[1], 'lon': row[2]} for row in data]
    print latlon
    
    #{'city': city, 'lat': lat, 'lon': lon for city, lat, lon in row} for row in data]
```

```python
for d in latlon:
    point = Point(float(d['lon']), float(d['lat']))
```

```python
schema = { 'geometry': 'Point', 'properties': { 'name': 'str' } }

with collection(
    "some.shp", "w", "ESRI Shapefile", schema) as output:
    for d in latlon:
        point = Point(float(d['lon']), float(d['lat']))
        output.write({
            'properties': {
                'name': d['city']
            },
            'geometry': mapping(point)
            })
```

```python
import csv
from nominatim import Nominatim
from collections import Counter
```

```python
f = 'usplayerswithsocNODUPE.csv'

with open(f, 'rt') as fin:
    fin = csv.reader(fin)
    headers = next(fin)
    data = [r for r in fin]
    
print headers
print len(data)
```

```python
lindex = headers.index('loc')

locs = [p[lindex] for p in data]

locdict = {loc: {'lat': None, 'lon': None} for loc in list(set(locs))[:100]}
```

```python
fails = []

for locname, latlon in locdict.iteritems():
    try:
        geolocator = Nominatim()
        info = geolocator.query(locname)[0]
        lat = info['lat']
        lon = info['lon']
        locdict[locname]['lat'] = lat
        locdict[locname]['lon'] = lon
        print "completed", locname
    except:
        fails.append(locname)
        print "failed on", locname
    
```

```python
print len(fails)
print fails
```

```python
llrows = []

for player in data:
    if int(player[2]) >=2014:
        continue
    row = []
    row.append(player[0])
    row.append(player[2] + '-01-01')
    row.append(player[2] + '-12-31')
    row.append(player[4])
    lat = latlons[player[4]]['lat']
    lon = latlons[player[4]]['lon']
    row.append(lat)
    row.append(lon)
    llrows.append(row)
```
